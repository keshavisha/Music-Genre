```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("features_30_sec.csv")
```

```python
df=df.drop("filename", axis="columns")
df=df.drop("length", axis="columns")
```

```python
df_blues=df.iloc[:100,43]
df_classical=df.iloc[100:200,43]
df_country=df.iloc[200:300,43]
df_disco=df.iloc[300:400,43]
df_hiphop=df.iloc[400:500,43]
df_jazz=df.iloc[500:600,43]
df_metal=df.iloc[600:700,43]
df_pop=df.iloc[700:800,43]
df_reggae=df.iloc[800:900,43]
df_rock=df.iloc[900:,44]
plt.scatter([1]*(len(df_blues)),df_blues)
plt.scatter([2]*(len(df_classical)),df_classical)
plt.scatter([3]*(len(df_country)),df_country)
plt.scatter([4]*(len(df_disco)),df_disco)
plt.scatter([5]*(len(df_hiphop)),df_hiphop)
plt.scatter([6]*(len(df_jazz)),df_jazz)
plt.scatter([7]*(len(df_metal)),df_metal)
plt.scatter([8]*(len(df_pop)),df_pop)
plt.scatter([9]*(len(df_reggae)),df_reggae)
plt.scatter([10]*(len(df_rock)),df_rock)
plt.xlabel("Labels")
plt.ylabel("mfcc14_mean")
plt.title("mfcc14_mean plots")
plt.show()
```
